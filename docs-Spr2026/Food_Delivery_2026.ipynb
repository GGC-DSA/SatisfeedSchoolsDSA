### Dataset 2

Link to website containing original Dataset: https://www.fns.usda.gov/pd/supplemental-nutrition-assistance-program-snap

This dataset was obtained from the Food and Nutrition Service and contains information on SNAP benefits across the USA. The data covers all substates/regions across the USA and provides information on the number of participants and households receiving SNAP benefits. In addition, both persons and households are split into groups of â€œPublic Assistanceâ€ and â€œNon-Public Assistanceâ€ participation.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df_snap = pd.read_excel('SNAP - National Data Bank JUL 2023.xlsx', header=2, skiprows=[0])
df_snap.head()
```

```python
df_snap = df_snap.dropna(axis=1, how='all')
df_snap.head()
```

```python
df_snap = df_snap[df_snap['Substate/Region'].astype(str).str.startswith('13')]
df_snap.head(n=10)
```

```python
# Extract identification code and insert at the beginning
df_snap.insert(0, 'identification code', df_snap['Substate/Region'].astype(str).str[:7])
df_snap.head()
```

```python
# Extract 'County' data
county_data = df_snap['Substate/Region'].astype(str).str.split(' EBT ', n=1).str[1].str.strip()
df_snap.insert(1, 'County', county_data)

df_snap.head()
```

```python
df_snap = df_snap.drop(columns=['Substate/Region'])
df_snap.head()
```

Although the dataset does not contain any missing data, it needed to be cleaned for only the relevant data. The first change I made was to remove all rows that did not contain information regarding Georgia. The Second change was to drop the empty columns. Although keeping all region-identifying data in a single column might have been beneficial, it was too clustered and contained unnecessary information. As a result, I took the relevant information (identification code and County) and stored them in separate columns. Finally, I deleted the old column (Substate/Region).

Below is the final dataset:

```python
df_snap.head()
```

Run to download the cleaned dataset:

```python
from google.colab import files
df_snap.to_csv('cleaned_SNAP - National Data Bank JUL 2023.csv', index=False)
files.download('cleaned_SNAP - National Data Bank JUL 2023.csv')
```

### Dataset 3

Link to website containing original Dataset: https://datacenter.aecf.org/data/tables/678-households-with-children-receiving-benefits-from-the-supplemental-nutrition-assistance-program-snap


This dataset was obtained from the Annie E. Casey Foundation and is very simple. The dataset consists of location type, location, timeframe, data format, and the number of households with children receiving SNAP benefits.


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df_snap_house = pd.read_excel('Households with children receiving benefits from SNAP.xlsx')
df_snap_house.head(n=10)
```

```python
df_snap_house = df_snap_house.drop(columns="DataFormat")
df_snap_house.head(n=10)
```

```python
years_to_keep = [2024, 2023, 2022, 2021]
df_snap_house = df_snap_house[df_snap_house['TimeFrame'].isin(years_to_keep)]
df_snap_house.head(n=10)
```

This dataset is very minimalistic and clean, so it did not require any major changes. The only changes made were dropping the DataFormat, and all rows that were in the following timeframe: 2024, 2023, 2022, or 2021. Since our client does not need data from far back, only data from recent years was kept.

Run to download clean dataset:

```python
from google.colab import files
df_snap_house.to_csv('cleaned_Households with children receiving benefits from SNAP.csv', index=False)
files.download('cleaned_Households with children receiving benefits from SNAP.csv')
```
